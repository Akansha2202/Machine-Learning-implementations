```python
query_point = np.array([2,3])
plt.scatter(X[:,0],X[:,1],c=Y)
plt.scatter(query_point[0],query_point[1],color='red')
plt.show()
```

```python
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))

def knn(X,Y,query_point,k=5):
    vals = []
    m = X.shape[0]
    
    for i in range(m):
        d = dist(query_point,X[i])
        vals.append((d,Y[i]))
    
    vals = sorted(vals)
    vals = vals[:k]
    vals = np.array(vals)
    
    cnt_vals = np.unique(vals[:,1],return_counts=True)
    #print(cnt_vals)
    idx = cnt_vals[1].argmax()
    pred = cnt_vals[0][idx]
    
    return pred
    #return vals
```

```python
pred_value = knn(X,Y,query_point)
print(pred_value)
```
